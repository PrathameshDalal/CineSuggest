```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
import ast
```

```python
movies_metadata = pd.read_csv('bollywood_full_1950-2019.csv')
crew_data = pd.read_csv('bollywood_crew_data_1950-2019.csv')
crew = pd.read_csv('bollywood_crew_1950-2019.csv')
```

```python
movies_metadata = movies_metadata.merge(crew, on='imdb_id')
```

```python
movies_metadata.drop_duplicates(subset='imdb_id', keep='first', inplace=True)
```

```python
movies = movies_metadata[['imdb_id', 'original_title', 'genres', 'summary', 'actors', 'directors', 'story']]
```

```python
movies
```

```python
movies['genres'] = movies['genres'].str.split('|').apply(lambda x: [genre.strip() for genre in x])
# Display the result
movies
```

```python
movies.info()
```

```python
movies['actors'] = movies['actors'].apply(lambda x: x if pd.isna(x) else x.split('|'))
movies
```

```python
movies['summary'] = movies['summary'].apply(lambda x: x if pd.isna(x) else x.split())
movies['story'] = movies['story'].apply(lambda x: x if pd.isna(x) else x.split())
movies
```

```python
movies['summary'] = movies['summary'] + movies['story']
movies.drop('story', inplace=True, axis=1)
```

```python
movies
```

```python
movies['directors'] = movies['directors'].apply(lambda x: x if pd.isna(x) else x.split())
movies
```

```python
import pandas as pd

# Load main CSV with movie information
movies_main = movies

# Function to replace crew IDs with director names
def replace_crew_ids_with_names(crew_ids_list):
    return [crew_data.loc[crew_data['crew_id'] == crew_id, 'name'].values[0] if not crew_data.loc[crew_data['crew_id'] == crew_id].empty else 'Unknown' for crew_id in crew_ids_list]

# Apply the function to the 'directors' column
movies_main['directors'] = movies_main['directors'].apply(replace_crew_ids_with_names)
movies['directors'] = movies_main['directors']

```

```python
movies
```

```python
movies['combined_features'] = movies['summary'] + movies['actors'] + movies['genres'] + movies['directors']
```

```python
import pandas as pd

from sklearn.metrics.pairwise import cosine_similarity
import torch
```

```python
movies = movies.dropna(subset=['combined_features'])
movies = movies.dropna(subset=['actors'])
movies = movies.dropna(subset=['summary'])
# Convert tokens to strings
#movies['combined_features'] = movies['combined_features'].apply(lambda tokens: ' '.join(tokens) if tokens else '')
#movies['actors'] = movies['actors'].apply(lambda tokens: ' '.join(tokens) if tokens else '')
#movies['summary'] = movies['summary'].apply(lambda tokens: ' '.join(tokens) if tokens else '')
#movies['genres'] = movies['genres'].apply(lambda tokens: ' '.join(tokens) if tokens else '')
#movies['directors'] = movies['directors'].apply(lambda tokens: ' '.join(tokens) if tokens else '')
```

```python
df = movies[['imdb_id', 'original_title', 'combined_features', 'summary', 'actors', 'genres', 'directors']]
```

```python
df
```

```python
df['combined_features'] = df['combined_features'].apply(lambda x:" ".join(x))
```

```python
df['combined_features'] = df['combined_features'].apply(lambda x:x.lower())
```

```python
df
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
df['combined_features'] = df['combined_features'].apply(stem)
```

```python
df['crew'] = df['actors'] + df['directors']
df.drop(['actors', 'directors'],inplace = True,axis=1)
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transformvectors = cv.fit_transform(df['combined_features']).toarray()
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key= lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    l = []
    movie_index = df[df['original_title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key= lambda x:x[1])[1:11]
    for i in movies_list:
        print(df.iloc[i[0]].original_title)
```

```python
recommend('3 Idiots')
```
